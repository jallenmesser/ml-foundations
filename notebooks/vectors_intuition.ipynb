```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(w):
    x, y = w
    return w**2 + y**2

def gradient(w):
    x,y = w
    return np.array([2*x, 2*y])
```

```python
w = np.array([3.0, -4.0])

print("w:", w)
print("f(w):", f(w))
print("gradient:", gradient(w))
```

```python
w = np.array([3.0, -4.0])
learning_rate = 0.1 

g = gradient(w)
w_new = w - learning_rate * g

print("w:", w)
print("gradient:", g)
print("w_new:", w_new)
print("f(w):", f(w))
print("f(w_new):", f(w_new))
```

```python
# Initialize w at some starting point
w = np.array([3.0, -4.0])
learning_rate = 0.1

# Store history of w and f(w) for plotting
w_history = [w.copy()]
f_history = [f(w)]

# Run 15 steps of gradient descent
for step in range(15):
    grad = gradient(w)
    w = w - learning_rate * grad
    
    # Record the new values
    w_history.append(w.copy())
    f_history.append(f(w))

# Print final result
print("Final w:", w)
print("Final loss f(w):", f(w))
```

```python
# Initialize w at the starting point
w = np.array([3.0, -4.0])
learning_rate = 0.1

# Run 15 steps of gradient descent and print each step
for step in range(15):
    grad = gradient(w)
    w = w - learning_rate * grad
    
    # Print the step number, w, and f(w)
    print(f"Step {step + 1}: w = {w}, f(w) = {f(w)}")

# Print the final result
print("\nFinal w:", w)
print("Final loss f(w):", f(w))
```

Linear Regression Learning

```python
# Simple dataset
x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)

plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Simple Linear Data")
plt.show()
```

```python
# Initial guess for the slope (just a random start)
w = 0.0

# Define the model: y_pred = w * x
def predict(x, w):
    return w * x
```

```python
# Mean squared error loss function
def loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
# Gradient of the loss with respect to w
def gradient(x, y_true, w):
    y_pred = predict(x, w)
    return -2 * np.mean(x * (y_true - y_pred))
```

```python
# Training loop
w = 0.0
learning_rate = 0.01

for step in range(20):
    grad = gradient(x, y, w)
    w = w - learning_rate * grad
    
    current_loss = loss(y, predict(x, w))
    print(f"Step {step+1}: w = {w:.4f}, loss = {current_loss:.4f}")
```

```python
# Plot the data
plt.scatter(x, y, label="Data")

# Plot the model's prediction line
x_fit = np.linspace(0, 6, 100)
y_fit = predict(x_fit, w)
plt.plot(x_fit, y_fit, color='red', label=f"Fitted line: y={w:.2f}x")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
